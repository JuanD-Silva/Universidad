```
# Soluci√≥n propuesta
import random
import matplotlib.pyplot as plt
from IPython.display import display
import ipywidgets as widgets
import time

class Tablero:
    def __init__(self, tamano=5, hojas=10):
        self.tamano = tamano
        self.hojas = set()
        while len(self.hojas) < hojas:
            self.hojas.add((random.randint(0, tamano - 1), random.randint(0, tamano - 1)))
        self.out = widgets.HTML()
        display(self.out)

    def dibujar(self, agente):
        tablero = "<table border='1'>"
        for i in range(self.tamano):
            tablero += "<tr>"
            for j in range(self.tamano):
                contenido = ""
                if (j, i) in self.hojas:
                    contenido = "üçÇ"
                if agente.x == j and agente.y == i:
                    contenido = agente.emoticon
                tablero += f"<td style='width:50px;height:50px;text-align:center;font-size:30px;'>{contenido}</td>"
            tablero += "</tr>"
        tablero += "</table>"
        self.out.value = tablero

    def limpiar(self, x, y):
        if (x, y) in self.hojas:
            self.hojas.remove((x, y))
            return True
        return False
```

```
##Haz un agente aspiradora con los siguientes requisitos

## * El agente aspiradora puede percibir las hojas en cualquiera de las 4 casillas adyacentes (N, S, W, E).
##* El agente puede percibir los muros del tablero
##  * El agente aspiradora dispondr√° de 6 acciones:  moverse a cualquiera de las 4 casillas adyacentes (N, S, W, E), aspirar o no realizar ninguna acci√≥n.
##  * El agente obtendr√° un punto por cada cuadrado que se limpie.
##  * La posici√≥n inicial de la aspiradora deber√° generarse de manera aleatoria en el tablero.
##  * La aspiradora solo tiene energ√≠a para realizar 100 movimientos. Las acciones de aspirar y percibir las hojas no consumen energ√≠a.
##  * El tablero debe tener un tama√±o de 5 x 5.
##  * Se deben ubicar 10 hojas de manera aleatoria en el tablero.
class Agente:
    def __init__(self, tablero):
        self.x = random.randint(0, tablero.tamano - 1)
        self.y = random.randint(0, tablero.tamano - 1)
        self.emoticon = "ü§ñ"
        self.energia = 100
        self.puntaje = 0
        self.tablero = tablero

    def percibir(self):
        percepcion = {
            'N': (self.x, self.y - 1) in self.tablero.hojas if self.y > 0 else None,
            'S': (self.x, self.y + 1) in self.tablero.hojas if self.y < self.tablero.tamano - 1 else None,
            'W': (self.x - 1, self.y) in self.tablero.hojas if self.x > 0 else None,
            'E': (self.x + 1, self.y) in self.tablero.hojas if self.x < self.tablero.tamano - 1 else None,
        }
        return percepcion

    def mover(self, direccion):
        if self.energia > 0:
            if direccion == 'N' and self.y > 0:
                self.y -= 1
            elif direccion == 'S' and self.y < self.tablero.tamano - 1:
                self.y += 1
            elif direccion == 'W' and self.x > 0:
                self.x -= 1
            elif direccion == 'E' and self.x < self.tablero.tamano - 1:
                self.x += 1
            self.energia -= 1

    def aspirar(self):
        if self.tablero.limpiar(self.x, self.y):
            self.puntaje += 1

    def actuar(self):
        while self.energia > 0:
            percepcion = self.percibir()
            if self.tablero.limpiar(self.x, self.y):
                self.puntaje += 1
            elif True in percepcion.values():
                for dir, hoja in percepcion.items():
                    if hoja:
                        self.mover(dir)
                        break
            else:
                self.mover(random.choice(['N', 'S', 'W', 'E']))
            self.tablero.dibujar(self)
            time.sleep(0.1)
```

```
# Ejecutar simulaciones
resultados = []
for _ in range(50):
    tablero = Tablero()
    agente = Agente(tablero)
    agente.actuar()
    resultados.append(agente.puntaje)

# Graficar histograma
plt.hist(resultados, bins=20, edgecolor='black')
plt.xlabel("Puntaje obtenido")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n del puntaje en 50 simulaciones")
plt.show()
```
